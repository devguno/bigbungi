```python
import pandas as pd


channel =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/channelInfo.csv')
video =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/youtube/videoInfo.csv')
display(channel.head())
display(video.head())
```

```python
channel.info()
```

```python
channel.ct = pd.to_datetime(channel.ct)
```

```python
result = channel[channel.ct >= pd.to_datetime('2021-10-03')].sort_values(['channelname','ct']).drop_duplicates('channelname',keep='first')
result[['subcnt','channelname']]
```

```python
channel['channelname'].value_counts()
```

```python
bef= channel[channel.ct.dt.date.astype('str') == '2021-10-03'].drop_duplicates('channelname',keep='first')[['subcnt','channelname']]
aft = channel[channel.ct.dt.date.astype('str') == '2021-11-01'].drop_duplicates('channelname',keep='last')[['subcnt','channelname']]
```

```python
pd.concat([bef.reset_index(drop=True), aft.reset_index(drop=True)],axis=1)
```
